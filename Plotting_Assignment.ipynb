```python
# Plotting a histogram of the 'co2' column
plt.figure(figsize=(8, 6))
plt.hist(df['co2'], bins=10, color='pink', edgecolor='black')

# Adding labels and title
plt.title('Distribution of CO2 Emissions')
plt.xlabel('CO2 Emissions')
plt.ylabel('Frequency')

# Display the histogram
plt.show()
```

Interpretation: The histogram shows the distribution of carbon dioxide emissions across different vehicles. The x-axis shows CO2 emission ranges from 0 to 700, while the y-axis indicates the number of vehicles within each emission range. The histogram displays a bell-shaped symmetrical curve, with the highest frequency around the 300-400 CO2 emissions range, where over 350 vehicles fall, which suggests a normal distribution. The frequencies are tapering off evenly away from the center, and no outliers are apparent.


```python
# Create a histogram of 'fescore'
plt.figure(figsize=(8, 6))
plt.hist(df['feScore'], bins=10, color='purple', edgecolor='black')

# Add labels and title
plt.title('Distribution of Fuel Efficiency Score', fontsize=14)
plt.xlabel('Fuel Efficiency Score', fontsize=12)
plt.ylabel('Frequency', fontsize=12)

# Display the plot
plt.show()
```

Interpretation: The histogram shows a nearly normal distribution of fuel efficiency scores, with most vehicles scoring between 4 and 6. The x-axis shows the fuel economy score while the y-axis shows the number of vehicles within the fuel efficiency score ranges. The highest frequency occurs in the 5-6 range, where around 1,000 vehicles fall. This suggests that the majority of vehicles in the dataset have average fuel efficiency performance. The distribution is relatively symmetric, with fewer vehicles having low scores (below 3) or high scores (above 8), indicating that extreme fuel efficiency performances are less common. No significant outliers are visible.

```python
# Plot a histogram of 'ghgscore'
plt.figure(figsize=(8, 6))
plt.hist(df['ghgScore'], bins=5, color='violet', edgecolor='black')

# Add labels and title
plt.title('Distribution of GHG Scores', fontsize=14)
plt.xlabel('GHG Score', fontsize=12)
plt.ylabel('Frequency', fontsize=12)

# Display the plot
plt.show()
```

Interpretation: The x-axis shows the greenhouse gas emissions score while the y-axis shows the number of vehicles within the ghg score ranges. The GHG score histogram also demonstrates a normal distribution, peaking between scores of 5 and 6, where more than 1,600 vehicles are found. This suggests that a large portion of the vehicles produce moderate levels of greenhouse gas emissions. Similar to the fuel economy distribution, this one is symmetric, with a sharp decline in the number of vehicles as you move away from the center, particularly toward the higher and lower GHG scores. There are no notable outliers in this dataset.

```python
# function to remove all the non numerical columns
def remove_colums(df):
    """
    return dataset with no non numerical value
    :param df: input dataFrame
    :return: clean dataFrame
    """
    return df.select_dtypes(include=["number"])

data = remove_colums(df)
data.head()
```

```python
#Create heatmap
correlation_matrix = data.corr()
sns.heatmap(correlation_matrix, vmin=-1, vmax=1, cmap='vlag')

#Add title
plt.title('Correlation Heat Map')

#Display heatmap
plt.show()
```

Interpretation of the Heat Map

The correlation heatmap shows key relationships between the features, particularly focusing on the ghgScore. There is a strong negative correlation between ghgScore and both co2 emissions and engine displacement (displ). This means that vehicles with higher CO2 emissions or larger engines tend to have lower GHG scores. On the other hand, feScore (fuel efficiency score) is positively correlated with ghgScore, which suggests that vehicles with better fuel efficiency generally have higher GHG ratings.

Some features, like pv2 and pv4, have very weak or no correlation with ghgScore and other variables, indicating that they likely don't contribute much useful information for predicting the target. Given this, if I were building a model to predict ghgScore, I would drop the id column (since its just an identifier), as well as pv2, pv4, and year because they show little to no correlation and don't seem to add much value. This would help the model focus on more relevant features and reduce unnecessary complexity.
